```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('/Users/jasonpaik9/Documents/Thinkful/Unit 2/US_Deaths.csv')
```

```python
data_over_time_1 = data.loc[(data.CauseName == 'Heart disease') & (data.CauseName != 'All causes') & (data.State == 'United States')]
data_OT_1 = data_over_time_1.drop(['113_Cause_Name','Age_Adj_Death_Rate'],axis = 1)
data_OT_HD = data_OT_1
data_over_time_2 = data.loc[(data.CauseName == 'Stroke') & (data.CauseName != 'All causes') & (data.State == 'United States')]
data_OT_2 = data_over_time_2.drop(['113_Cause_Name','Age_Adj_Death_Rate'],axis = 1)
data_OT_ST = data_OT_2
data_over_time_3 = data.loc[(data.CauseName == 'Cancer') & (data.CauseName != 'All causes') & (data.State == 'United States')]
data_OT_3 = data_over_time_3.drop(['113_Cause_Name','Age_Adj_Death_Rate'],axis = 1)
data_OT_CA = data_OT_3
data_over_time_4 = data.loc[(data.CauseName == 'Unintentional injuries') & (data.CauseName != 'All causes') & (data.State == 'United States')]
data_OT_4 = data_over_time_3.drop(['113_Cause_Name','Age_Adj_Death_Rate'],axis = 1)
data_OT_UI = data_OT_4

plt.subplots(1,sharex=True)
plt.scatter(x=data_OT_HD.Year,y=data_OT_HD.Deaths,color='blue')
plt.plot(x=data_OT_HD.Year,y=data_OT_HD.Deaths,color='blue')
plt.title('Heart Disease Deaths from 1999-2015')
plt.xlabel('Year')
plt.xlim([1999,2015])
plt.ylabel('Deaths')

plt.subplots(1,sharex=True)
plt.scatter(x=data_OT_ST.Year,y=data_OT_ST.Deaths,color='red')
plt.plot(x=data_OT_ST.Year,y=data_OT_ST.Deaths,color='red')
plt.title('Stroke Deaths from 1999-2015')
plt.xlabel('Year')
plt.xlim([1999,2015])
plt.ylabel('Deaths')

plt.subplots(1,sharex=True)
plt.scatter(x=data_OT_CA.Year,y=data_OT_CA.Deaths,color='green')
plt.plot(x=data_OT_CA.Year,y=data_OT_CA.Deaths,color='green')
plt.title('Cancer Deaths from 1999-2015')
plt.xlabel('Year')
plt.xlim([1999,2015])
plt.ylabel('Deaths')

plt.subplots(1,sharex=True)
plt.scatter(x=data_OT_UI.Year,y=data_OT_UI.Deaths,color='purple')
plt.plot(x=data_OT_UI.Year,y=data_OT_UI.Deaths,color='purple')
plt.title('Unintentional Injuries Deaths from 1999-2015')
plt.xlabel('Year')
plt.xlim([1999,2015])
plt.ylabel('Deaths')

plt.show()

## In this graph, I did a simple plyplot.scatter to show the deaths between Heart Disease, Strokes,
## Cancer, and Unintentional Deaths. Heart disease, by the numbers, has the highest number of deaths
## but plotting the data year-to-year, HD deaths have regressed, along with strokes. Interestingly
## those two deaths are spiked since 2010. The correlations between cancer and unintentional deaths
## are also the same shape which is interesting not knowing the correlation between those two.
## All in all, there are medical questions I'm not quite sure of but these graphs tell an interesting
## narrative of death types that are "unrelated" to each other yet show similar graph patterns.
```

```python
data_sickness_1 = data.loc[(data.CauseName == 'Influenza and pneumonia') & (data.CauseName != 'All causes') & (data.State == 'United States')]
data_sick_1 = data_sickness_1.drop(['113_Cause_Name'],axis = 1)
data_sick_1 = data_sick_1.sort_values(by=['Year'])

plt.subplots(1,sharex=True)
plt.plot(data_sick_1['Year'],data_sick_1['Deaths'])
plt.title('# of Flu/Pneumonia Deaths from 1999-2015')
plt.xlabel('Year')
plt.xlim([1999,2015])
plt.ylim([47000,70000])
plt.ylabel('Deaths')

plt.subplots(1,sharex=True)
plt.plot(data_sick_1['Year'],data_sick_1['Age_Adj_Death_Rate'],color='navy')
plt.title('Avg Age Adjusted Death Rates of Flu/Pneumonia Deaths from 1999-2015')
plt.xlabel('Year')
plt.xlim([1999,2015])
plt.ylim([13.5,27])
plt.ylabel('Deaths')
plt.show()

## In the data, there were two death types that were "preventable" (i.e. compared to Stroke, Cancer, etc.) and I 
## did a pyplot.plot graph between influenza and pneumonia (grouped as one column in dataset), to curiously see
## their trends across overall deaths and age adjusted death rates, a metric to qualify age and old age.
## The shape of the graphs are similar but the age adjusted death rates graph shows us that over the last 
## 16 years elderly people have experienced less deaths from flu/pneumonia whether that be to better 
## medicine and better treatment options for these types of illnesses.
```

```python
data_southern = data[data['State'].isin(['Delaware','Florida','Georgia','Maryland','North Carolina','South Carolina',
                            'Virginia','West Virginia','Alabama','Kentucky','Mississippi','Tennessee',
                            'Arkansas','Louisiana','Oklahoma','Texas'])
         & data['CauseName'].isin(['Diabetes','Suicide'])]
data_midwest = data[data['State'].isin(['Illinois','Indiana','Iowa','Kansas','Michigan','Minnesota','Missouri',
                            'Nebraska','North Dakota','Ohio','South Dakota','Wisconsin'])
         & data['CauseName'].isin(['Diabetes','Suicide'])]
data_northeast = data[data['State'].isin(['Maine','New York','New Jersey','Vermont','Massachusetts','Rhode Island',
                            'New Hampshire','Pennsylvania'])
         & data['CauseName'].isin(['Diabetes','Suicide'])]
data_western = data[data['State'].isin(['Montana','Wyoming','Colorado','New Mexico','Idaho','Utah','Arizona','Nevada',
                            'Washington','Oregon','California','Alaska','Hawaii'])
         & data['CauseName'].isin(['Diabetes','Suicide'])]
```

```python
data_southern_suicide = data_southern.loc[(data_southern.CauseName == 'Suicide')]
data_southern_suicide = data_southern_suicide.groupby("Year").mean().reset_index()
data_southern_suicide = data_southern_suicide.sort_values(by=['Year'])
data_midwest_suicide = data_midwest.loc[(data_midwest.CauseName == 'Suicide')]
data_midwest_suicide = data_midwest_suicide.groupby("Year").mean().reset_index()
data_midwest_suicide = data_midwest_suicide.sort_values(by=['Year'])
data_northeast_suicide = data_northeast.loc[(data_northeast.CauseName == 'Suicide')]
data_northeast_suicide = data_northeast_suicide.groupby("Year").mean().reset_index()
data_northeast_suicide = data_northeast_suicide.sort_values(by=['Year'])
data_western_suicide = data_western.loc[(data_western.CauseName == 'Suicide')]
data_western_suicide = data_western_suicide.groupby("Year").mean().reset_index()
data_western_suicide = data_western_suicide.sort_values(by=['Year'])

plt.subplots(1,sharex = True)
plt.plot(data_southern_suicide.Year,data_southern_suicide.Deaths,color='blue',label = 'Southern')
plt.plot(data_midwest_suicide.Year,data_midwest_suicide.Deaths,color='purple',label = 'Midwest')
plt.plot(data_northeast_suicide.Year,data_northeast_suicide.Deaths,color='red',label = 'Northeast')
plt.plot(data_western_suicide.Year,data_western_suicide.Deaths,color='green',label = 'Western')
plt.legend(loc = 'upper left')
plt.title('Geographical Suicide Deaths from 1999-2015')
plt.xlabel('Year')
plt.xlim([1999,2015])
plt.ylabel('Deaths')
plt.show()

## For the next two graphs, I re-organized the data to four different subgroups - southern, midwest, northeast
## and western - all of which are four geographic territories in the US as defined by the US Census Bureau. 
## The overall populations of these four territories will sway these graphs but the graph below shows year
## to year suicides in the US, all of which are rising rapidly since 1999. Sharp upticks in suicides in the 
## Northeast and Western line plots have noticeably risen while Southern states have also risen as well -
## we can maybe attest this to not just population growth but other sociological issues such as rising mental
## illnesses, etc.
```

```python
data_southern_diabetes = data_southern.loc[(data_southern['CauseName'] == 'Diabetes') & (data_southern['Year'] == 2016)]
data_midwest_diabetes = data_midwest.loc[(data_midwest['CauseName'] == 'Diabetes') & (data_midwest['Year'] == 2016)]
data_western_diabetes = data_western.loc[(data_western['CauseName'] == 'Diabetes') & (data_western['Year'] == 2016)]
data_northeast_diabetes = data_northeast.loc[(data_northeast['CauseName'] == 'Diabetes') & (data_northeast['Year'] == 2016)]

plt.scatter(data_southern_diabetes.Age_Adj_Death_Rate,data_southern_diabetes.Deaths,color='blue',label = 'Southern')
plt.scatter(data_midwest_diabetes.Age_Adj_Death_Rate,data_midwest_diabetes.Deaths,color='purple',label = 'Midwest')
plt.scatter(data_northeast_diabetes.Age_Adj_Death_Rate,data_northeast_diabetes.Deaths,color='red',label = 'Northeast')
plt.scatter(data_western_diabetes.Age_Adj_Death_Rate,data_western_diabetes.Deaths,color='green',label = 'Western')
plt.legend(loc = 'upper right')
plt.title('Geographical Diabetes Deaths in 2016')
plt.xlabel('Year')
plt.ylabel('Deaths')
plt.show()

## In this scatter plot, I used the same variables above using US Geographical segments to plot diabetes deaths
## in 2016 with the age adjusted death rates on the x-axis and the overall deaths on the y-axis. Each point
## in the scatter plot represents a state and its average age death rate per diabetes deaths. We should see 
## higher deaths correlate to populations but interestingly we see more elderly people with diabetes deaths
## in Southern states in the bottom right. We can most likely attribute this to dietary lifestyles seeing
## fewer diabetes deaths in places (bottom left) to places like Hawaii.
```
